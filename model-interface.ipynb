```python
import gradio as gr
import tensorflow as tf
print("moudules imported")
```

```python
model = tf.keras.models.load_model("/kaggle/input/minor-model-2/conv2d-minor project model-96.00.h5")
```

```python
def image_classifier(image):
#     Resize image to (224, 224) and expand dimensions to (1, 224, 224, 3)
    image = tf.image.resize(image, (256, 256))
    image = tf.keras.preprocessing.image.img_to_array(image)
    image = tf.expand_dims(image, axis=0)
    labels = {0:'Late_blight',1:'healthy',2:'Early_blight',3:'Septoria_leaf_spot',4:'Tomato_Yellow_Leaf_Curl_Virus',5:'Bacterial_spot', 6:'Target_Spot', 7:'Tomato_mosaic_virus', 8:'Leaf_Mold', 9:'Spider_mites Two-spotted_spider_mite'}
    prediction = model.predict(image)
#     return prediction
    return labels[prediction.argmax()][:-1]
```

```python
demo = gr.Interface(fn=image_classifier, inputs="image", outputs="label")
demo.launch()
```
